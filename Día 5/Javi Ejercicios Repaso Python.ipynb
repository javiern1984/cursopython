## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
contador = 0
#Respuesta:
#Debe dar 5
for iterar in nums:
  contador = contador+1
  if iterar == 0:
      break
print(contador)

```

## Ejercicio 3

```python
nums = [6,8,5,2,0,6,5]
suma = 0
contador = 0
#Respuesta:
#Debe dar 4.2
for iterar in nums:
  contador = contador+1
  suma = suma + iterar
  media = suma / contador
  if iterar == 0:
    break
print(media)
```

## Ejercicio 5 (Opcional)

Crear una lista a partir de una secuencia de valores de entrada terminada en *.
Una vez creada, modificarla elevando al cuadrado todos los pares y al cubo todos los impares, y mostrarla.

```python
lista = [3,56,6,7,"*",2,78]
new_lista = []
#Respuesta:
#Debe dar [27, 3136, 36, 343]
for iterar in lista:
  if iterar == "*":
    break
  if iterar %2 == 0:
    iterar = iterar ** 2
  if iterar %2 != 0:
    iterar = iterar ** 3
  new_lista.append(iterar)
print(new_lista)
```
