# Ejercicio 4

Aplicar las operaciones aritméticas básicas sobre los objetos Series dados.

Prueba a:
* Sumar
* Restar
* Multiplicar
* Dividir

````python
serie1 = pd.Series([1, 2, 3, 4, 5])
serie2 = pd.Series([9, 8, 7, 6, 5])
````

````python
def suma (a, b):
  resultado = (a + b)
  print(resultado)
suma(serie1, serie2)

def resta (a, b):
  resultado2 = (a - b)
  print(resultado2)
resta(serie1, serie2)

def multi (a, b):
  resultado3 = (a * b)
  print(resultado3)
multi(serie1, serie2)

def divi (a, b):
  resultado4 = (a / b)
  print(resultado4)
divi(serie1, serie2)
````

# Ejercicio 5

Usar operadores relacionales para comparar los objetos Series dados.



````python
serie1 = pd.Series([1, 2, 3, 5, 7])
serie2 = pd.Series([13, 17, 11, 5, 3])
````

````python
def igual (a, b):
  sonigual = (a == b)
  print(sonigual)
igual(serie1, serie2)

def mayorque (a, b):
  esmayor = (a > b)
  print(esmayor)
mayorque(serie1, serie2)

def menoroigualque (a, b):
  esmenoroigual = (a <= b)
  print(esmenoroigual)
menoroigualque(serie1, serie2)
````
